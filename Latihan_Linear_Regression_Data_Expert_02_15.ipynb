```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#dummy data jumlah kamar
bedrooms = np.array([1,1,2,2,3,4,4,5,5,5])

house_price = np.array ([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000, 85000, 90000])
```

```python
#menampilkan scatter plot dataset
plt.scatter(bedrooms, house_price)
plt.xlabel("Jumlah Kamar")
plt.ylabel("Harga Rumah")
plt.show()
```

```python
#import lib utk linear regression
from sklearn.linear_model import LinearRegression
```

```python
#melatih model dg algoritma
bedrooms = bedrooms.reshape(-1,1)
model = LinearRegression()
model.fit(bedrooms, house_price)
```

```python
#visualisasi hubungan antara jml kamar dg harga rumah

plt.scatter(bedrooms, house_price)
plt.plot(bedrooms, model.predict(bedrooms), color='red')
plt.xlabel("Jumlah Kamar")
plt.ylabel("Harga Rumah")
plt.show()
```
