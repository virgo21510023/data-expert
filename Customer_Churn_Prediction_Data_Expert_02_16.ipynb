###Menghapus Unnecessary Columns dari data

```python
cleaned_df = df_load.drop(['customerID'], axis=1)
print(cleaned_df.head())
```

###Encoding Data

```python
for column in cleaned_df.columns:
    if cleaned_df[column].dtype == np.number:
        continue # This line should be indented to be part of the if statement
    cleaned_df[column] = LabelEncoder().fit_transform(cleaned_df[column]) # Correct indentation
    print(cleaned_df.describe()) # Correct indentation
```

###Splitting Dataset

```python
x = cleaned_df.drop('Churn', axis=1)
y = cleaned_df['Churn']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

print('Jumlah baris dan kolom dari x_train adalah', x_train.shape)
print('Jumlah baris dan kolom dari y_train adalah', y_train.shape)
print('Prosentase Churn di data training adalah', y_train.value_counts(normalize=True))
print('Jumlah baris dan kolom dari x_test adalah', x_test.shape)
print('Jumlah baris dan kolom dari y_test adalah', y_test.shape)
print(y_test.value_counts(normalize=True))
```

###Pembuatan Model

```python
log_model = LogisticRegression()
log_model.fit(x_train, y_train)

y_pred = log_model.predict(x_test)
print(classification_report(y_test, y_pred))

print('Model Logistic Regression yang terbentuk adalah:')
print(log_model)
```
